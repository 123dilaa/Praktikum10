```python
import pandas as pd
import matplotlib.pyplot as plt 
# data 
jurusan = ['Ilmu Komputer', 'Sistem Informasi', 'Teknik Informatika', 'Data Science']
jumlah_mahasiswa = [120, 150, 100, 80]

# data tambahan untuk tahun yang berbeda
jurusan_mahasiswa_2024 =[110, 140, 95, 20]

x = range(len(jurusan)) #index posisi batang
width = 0.4

plt.bar(x, jumlah_mahasiswa, width=width, label='2023', color='skyblue')
plt.bar([p + width for p in x], jurusan_mahasiswa_2024, width=width, label='2024', color='yellow')

plt.title('Jumlah mahasiswa per jurusan (2023-2024)')
plt.xlabel('Jurusan')
plt.ylabel('jumlah_mahasiswa')
plt.xticks(rotation=45)
plt.xticks([p + width / 2 for p in x], jurusan) # mengatur label x-axis di tengah
plt.legend()
plt.show()
```
