```python
import pandas as pd
import matplotlib.pyplot as plt

#data
# data 
jurusan = ['Ilmu Komputer', 'Sistem Informasi', 'Teknik Informatika', 'Data Science']
jumlah_mahasiswa = [140, 150, 10, 99]

#menambahkan kostumisasi
colors = ['blue', 'yellow', 'red', 'purple']
plt.bar(jurusan, jumlah_mahasiswa, color=colors)
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=50)
for i, v in enumerate(jumlah_mahasiswa):
    plt.text(i, v + 3, str(v), color='black', ha='center') # menambahkan label pada batang
plt.show()
```
